```python
# load packages for API calls and data manipulation
import os
import requests
import json
import pandas as pd
from datetime import datetime
import time
import logging
from typing import Optional
import base64, json

# Must be imported BEFORE any Azure SDK or requests usage
import pip_system_certs.wrapt_requests

# load packages for Azure Key Vault to get API key
from azure.keyvault.secrets import SecretClient
from azure.identity import DefaultAzureCredential
```

```python
# set vault URI and key name
vault = "https://tdg-kv.vault.azure.net/"
key_name = "USPTO-ApiKey"

# Use the SAME credential your Key Vault client uses
credential = DefaultAzureCredential()

# Debug: inspect which identity is being used
tok = credential.get_token("https://vault.azure.net/.default").token
payload = tok.split(".")[1] + "=="
claims = json.loads(base64.urlsafe_b64decode(payload))
print("appid:", claims.get("appid"))
print("oid:", claims.get("oid"))
print("upn:", claims.get("upn"))
print("preferred_username:", claims.get("preferred_username"))
print("tid:", claims.get("tid"))

# Then your existing code
client = SecretClient(vault_url=vault, credential=credential)
secret = client.get_secret(key_name)

print("Secret value:", secret.value)
```
